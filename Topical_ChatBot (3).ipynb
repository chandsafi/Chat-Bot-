```python
def preprocess_data(data, intent=None):
    processed_data = []
    for entry in data:
        message = entry.get("message") or entry.get("user_query")
        response = entry.get("response") or entry.get("reply")

        if message and response:
            message_tokens = tokenizer.encode(message, truncation=True, padding='max_length')
            response_tokens = tokenizer.encode(response, truncation=True, padding='max_length')

            # Only add if tokens are not empty
            if message_tokens and response_tokens:
                processed_data.append({
                    "input_text": message,
                    "input_tokens": message_tokens,
                    "output_text": response,
                    "output_tokens": response_tokens,
                    "intent": intent
                })
    return processed_data
```
